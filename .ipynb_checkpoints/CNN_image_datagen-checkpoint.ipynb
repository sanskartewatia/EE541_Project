```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import torchaudio
import librosa
import os
import random
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import torch.nn.functional as F
from torch.utils.data import TensorDataset
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.signal import spectrogram
```

```python
sample_rate, samples = wavfile.read('./af/Major/Major_100.wav')
segment_duration = 1.75

# Compute the number of samples in the segment
segment_length = int(segment_duration * sample_rate)

# Extract the first segment_length samples of the audio signal
segment = samples[:segment_length]
frequencies, times, spectrogram_data = spectrogram(segment, fs=sample_rate,nfft = 1024)



# Plot the zoomed-in version
plt.pcolormesh(times, frequencies[frequencies <= 1000], 
               spectrogram_data[frequencies <= 1000, :])
plt.show()
```

```python
input_dir = './af/Minor'
output_dir = './pics/Minor'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

```python
image_width = 400
image_height = 300
for filename in os.listdir(input_dir):
    if filename.endswith('.wav'):
        # Load the audio data
        filepath = os.path.join(input_dir, filename)
        sample_rate, samples = wavfile.read(filepath)

        segment_length = int(1.75 * sample_rate)

        segment = samples[:segment_length]
        # Compute the spectrogram
        frequencies, times, spectrogram_data = spectrogram(segment, fs=sample_rate,nfft = 1024)

        # Plot the spectrogram
        fig, ax = plt.subplots(figsize=(image_width/100, image_height/100))
        ax.pcolormesh(times, frequencies[frequencies <=1000], spectrogram_data[frequencies <= 1000, :], cmap='hot')
        ax.axis('off')

        # Save the spectrogram as a JPEG file
        output_filename = os.path.splitext(filename)[0] + '.jpg'
        output_filepath = os.path.join(output_dir, output_filename)
        plt.savefig(output_filepath, dpi=100, bbox_inches='tight', pad_inches=0)

        # Clear the current plot to prepare for the next file
        plt.clf()
```
